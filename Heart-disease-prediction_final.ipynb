```python
#check for missing values
df.isna().sum() 
```

```python
#impute missing values

# Calculate the median of 'education' and 'BPmeds' columns
median_educ = df['education'].median()
median_BPMeds = df['BPMeds'].median()

# Fill missing values with the respective medians
df['education'].fillna(median_educ, inplace=True)
df['BPMeds'].fillna(median_BPMeds, inplace=True)

# Calculate the meam of 'cigsPerDay', 'totChol', 'BMI', 'heartrate' and 'glucose' columns
mean_cigsPerDay = df['cigsPerDay'].mean()
mean_totChol = df['totChol'].mean()
mean_BMI = df['BMI'].mean()
mean_heartrate = df['heartRate'].mean()
mean_glucose = df['glucose'].mean()

# Fill missing values with the respective means
df['cigsPerDay'].fillna(mean_cigsPerDay, inplace=True)
df['totChol'].fillna(mean_totChol, inplace=True)
df['BMI'].fillna(mean_BMI, inplace=True)
df['heartRate'].fillna(mean_heartrate, inplace=True)
df['glucose'].fillna(mean_glucose, inplace=True)
```

```python
#standardisation
columns= ['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
data = df[columns]

# Standardize the continuous variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Replace the original continuous variables with the standardized ones
df[columns] = scaled_data
print(df.head)
```

```python
#hot encoding education
encoded_df = pd.get_dummies(df['education'], prefix='education')
df_new = pd.concat([df.iloc[:,:2],encoded_df, df.iloc[:,3:]], axis=1)
print(df_new.head)
```

```python
#train-test split
from sklearn.model_selection import train_test_split
y = df_new.iloc[:,18]
X = df_new.iloc[:,0:18]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier

dt =DecisionTreeClassifier(criterion = 'entropy', random_state=0)
dt.fit(X_train, y_train)

#Accuracy
print("Accuracy is:", dt.score(X_test, y_test) )


#plot the features and feature_importances without zero and sorted
_
my_id = dt.feature_importances_>0
mynew_x =X_train.columns[my_id]
mynew_y = dt.feature_importances_[my_id]

#to sort the data and get them in decreasing order
sorted_data = sorted(zip(mynew_y, mynew_x), reverse=False)
mynew_y, mynew_x = zip(*sorted_data)

#confusion matrix
from sklearn.metrics import confusion_matrix

y_pred_dt = dt.predict(X_test)

my_matrix = confusion_matrix(y_test, y_pred_dt)
print(my_matrix)

#roc curve
from sklearn.metrics import roc_auc_score

dt_prob = dt.predict_proba(X_test)[:, 1]

print("AUC for the model:",  roc_auc_score(y_test, dt_prob) )

print('the importance of features is as follows:')
plt.barh(mynew_x, mynew_y)
```

```python
#removing diabetes and currebtSmoker from the attributes since they have least importance and also they are correlated with glucose and cigsPerDay
X_new = X.drop(['diabetes', 'currentSmoker'], axis=1)

X_train_n, X_test_n, y_train, y_test = train_test_split(X_new, y, test_size = 0.3, random_state=42)
```

```python
#also adding Gridsearch to the Decision tree model

from sklearn.model_selection import GridSearchCV

opt_tree = DecisionTreeClassifier(random_state = 0)

dt_params = {
    'max_depth': range(2,11),
    'min_samples_split': range(2,11),
    'min_samples_leaf': range(2,11),
    'criterion': ['gini', 'entropy']
}
 
grid_tree = GridSearchCV(opt_tree, dt_params)
grid_tree.fit(X_train_n, y_train)

print(grid_tree.best_params_)

#Accuracy
print("Accuracy is:", grid_tree.score(X_test_n, y_test) )

#confusion matrix
from sklearn.metrics import confusion_matrix

y_pred_gdt = grid_tree.predict(X_test_n)

my_matrix = confusion_matrix(y_test, y_pred_gdt)
print(my_matrix)

#roc curve
from sklearn.metrics import roc_auc_score

gdt_prob = grid_tree.predict_proba(X_test_n)[:, 1]

print("AUC for the model:",  roc_auc_score(y_test, gdt_prob) )
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
fig, ax = plt.subplots(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred_gdt)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=grid_tree.classes_)
name = 'decision tree'
disp.plot(cmap=plt.cm.Blues, ax=ax)
ax.title.set_text(f'Confusion Matrix for {name}')
ax.set_xlabel('Predicted Labels')
ax.set_ylabel('True Labels')

plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Get the best estimator from the grid search
best_tree = grid_tree.best_estimator_

# Convert DataFrame columns to a list
feature_names_list = X_train_n.columns.tolist()

# Plotting the decision tree with annotations
plt.figure(figsize=(20, 10))
plot_tree(best_tree, filled=True, feature_names=feature_names_list, class_names=['No Disease', 'Disease'], rounded=True, proportion=False, precision=2)
plt.title('Decision Tree - Best Estimator from Grid Search')
plt.show()
```

```python
# Linear SVC with GridSearchCV
from sklearn.svm import LinearSVC

# Define Function
linear_svc = LinearSVC(random_state = 22, max_iter = 20000) # must specify random state here

# Define a list of hyperparameters
params_svc = {'C': [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 100000]  }

grid_lr_svc = GridSearchCV(linear_svc, params_svc, n_jobs=2)

grid_lr_svc.fit(X_train_n, y_train)

# Print the best parameters found
print("Best Parameters:", grid_lr_svc.best_params_)
#Accuracy
print("Accuracy is:", grid_lr_svc.score(X_test_n, y_test) )

#confusion matrix
from sklearn.metrics import confusion_matrix

y_pred_lrsvc = grid_lr_svc.predict(X_test_n)

my_matrix = confusion_matrix(y_test, y_pred_lrsvc)
print(my_matrix)

#roc curve
from sklearn.metrics import roc_auc_score

lrsvc_prob = grid_lr_svc.decision_function(X_test_n)

print("AUC for the model:",  roc_auc_score(y_test, lrsvc_prob) )

```

```python
#Evaluate the voting classifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, auc
# Dictionary to hold name, predictions, and probabilities for each model
models = {
    "SVM": grid_lr_svc,
    "Logistic Regression": logreg,
    "Random Forest": grid_rnd,
    "Voting Classifier (Soft)": voting_clf
}

plt.figure(figsize=(10, 8))
for name, model in models.items():
    model.fit(X_train_n, y_train)
    y_pred = model.predict(X_test_n)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} accuracy: {accuracy:.4f}")

    # Check if the model can estimate probabilities (SVM and Voting Classifier can)
    if hasattr(model, "decision_function"):
        y_score = model.decision_function(X_test_n)
        # Convert decision function scores to probabilities
        y_prob = (y_score - y_score.min()) / (y_score.max() - y_score.min())
    elif hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test_n)[:, 1]
    else:
        continue  # Skip ROC curve if probabilities are not available

    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)
    print(f"{name} ROC AUC: {roc_auc:.4f}")

    # Plot ROC curve
    plt.plot(fpr, tpr, label=f'{name} (area = {roc_auc:.4f})')

# Plot ROC curve for a random classifier as a baseline
plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc="lower right")
plt.show()

```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
axes = axes.flatten()

for ax, (name, model) in zip(axes, models.items()):
    model.fit(X_train_n, y_train)
    y_pred = model.predict(X_test_n)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=model.classes_)
    disp.plot(cmap=plt.cm.Blues, ax=ax)
    ax.title.set_text(f'Confusion Matrix for {name}')
    ax.set_xlabel('Predicted Labels')
    ax.set_ylabel('True Labels')

plt.tight_layout()
plt.show()
```
